```python
import matplotlib.pyplot as plt
import numpy as np

x_min = -10
x_max = 10
y_min = -10
y_max = 10

# Data
points = 2 * (x_max - x_min)
x = np.linspace(x_min, x_max, points)
y1 = 2 * x + 1
y2 = 2 * (x ** 2) + 1
y3 = 2 * (x ** 3) + 1

# Figure
fix, ax = plt.subplots()
plt.axis([x_min, x_max, y_min, y_max])

# Axis
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_title('Polynomial functions')
ax.grid(True)
ax.set_xticks(np.arange(x_min, x_max + 1, 1))
ax.set_yticks(np.arange(y_min, y_max + 1, 1))

# Plot
plt.plot([x_min, x_max], [0,0], 'b')
plt.plot([0,0], [y_min, y_max], 'b')
plt.plot(x, y1, label="y = 2x + 1")
plt.plot(x, y2, label="y = 2x^2 + 1")
plt.plot(x, y3, label="y = 2x^3 + 1")
plt.plot([5],[4], 'ro', label="Point")

ax.legend()
plt.show()
```
