# Graphing Systems

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import var, Eq
from sympy.solvers import solve


# Graphing essentials
x_min, y_min, x_max, y_max = -10, -10, 10, 10
fig, ax = plt.subplots()
plt.axis([x_min, x_max, y_min, y_max])
plt.plot([x_min, x_max], [0,0], 'b')
plt.plot([0,0], [y_min, y_max], 'b')

# Solve equations
var('z')
lhs = z ** 2
rhs = 2 * z
eq = Eq(lhs, rhs)
solutions = solve(eq, z)
for solution in solutions:
  f1 = solution ** 2
  f2 = 2 * solution
  plt.plot([solution], [f1], 'ro')
  plt.annotate(f'({solution}, {f1})', xy=(solution, f1), xytext=(solution+0.5, f1))
  plt.plot([solution], [f2], 'ro')
  plt.annotate(f'({solution}, {f2})', xy=(solution, f2), xytext=(solution+0.5, f2))


# Axis settings
ax.set_xticks(np.arange(x_min, x_max + 1, 1))
ax.set_yticks(np.arange(y_min, y_max + 1, 1))
ax.grid(True)
ax.set_title("Linear and polynomial equations")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Data
x = np.linspace(x_min, x_max, 10*(x_max - x_min))
y1 = x**2
y2 = 2 * x

# Graph the equations
plt.plot(x, y1, label="y1= x^2")
plt.plot(x, y2, label="y2 = 4x")


ax.legend()
plt.show()
```

# Cool extra features

Fill between

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import var, Eq
from sympy.solvers import solve


# Graphing essentials
x_min, y_min, x_max, y_max = -10, -10, 10, 10
fig, ax = plt.subplots()
plt.axis([x_min, x_max, y_min, y_max])
plt.plot([x_min, x_max], [0,0], 'b')
plt.plot([0,0], [y_min, y_max], 'b')

# Axis settings
ax.set_xticks(np.arange(x_min, x_max + 1, 1))
ax.set_yticks(np.arange(y_min, y_max + 1, 1))
ax.grid(True)
ax.set_title("Linear equations using fill between graph")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Data
x = np.linspace(x_min, x_max, 10*(x_max - x_min))

# Graph the equations

# Line 3
y3 = 2 * x + 10
plt.plot(x, y2, label="y3 = 2x + 10")
plt.fill_between(x, y3, y_max, facecolor='blue')

# Line 2
y2 = 2 * x + 5
plt.plot(x, y2, label="y2 = 2x + 5")
plt.fill_between(x, y2, y3, facecolor='orange')

# Line 1
y1 = 2 * x
plt.plot(x, y1, label="y1= 2x")
plt.fill_between(x, y1, y2, facecolor='green')

ax.legend()
plt.show()
```

# Interactive Graphs

Interactive Graphs

```python
%matplotlib inline
from ipywidgets import interactive
import numpy as np
import matplotlib.pyplot as plt
from sympy import var, Eq
from sympy.solvers import solve

def func(m, b, zoom):
  # Graphing essentials
  x_min, y_min, x_max, y_max = -zoom, -zoom, zoom, zoom
  fig, ax = plt.subplots()
  plt.axis([x_min, x_max, y_min, y_max])
  plt.plot([x_min, x_max], [0,0], 'b')
  plt.plot([0,0], [y_min, y_max], 'b')

  # Solve equations
  var('z')
  lhs = z ** 2
  rhs = m * z + b
  eq = Eq(lhs, rhs)
  solutions = solve(eq, z)
  for solution in solutions:
    f1 = solution ** 2
    f2 = m * solution + b
    plt.plot([solution], [f1], 'ro')
    plt.plot([solution], [f2], 'ro')


  # Axis settings
  ax.grid(True)
  ax.set_title("Linear and polynomial equations")
  ax.set_xlabel("x")
  ax.set_ylabel("y")

  # Data
  x = np.linspace(x_min, x_max, 10*(x_max - x_min))
  y1 = x**2
  y2 = m * x + b

  # Graph the equations
  plt.plot(x, y1, label="y1= x^2")
  plt.plot(x, y2, label=f"y2 = {m}x + {b}")


  ax.legend()
  plt.show()

interactive_graph = interactive(func, m=(-10, 10), b=(-10, 10), zoom=(1, 100))
interactive_graph
```
