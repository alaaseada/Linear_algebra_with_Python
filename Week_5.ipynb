# Slope

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols
from sympy.solvers import solve

# Calculate slope and intercept
x1, y1, x2, y2 = 0, 875, 6, 575
m = (y2 - y1) / (x2 - x1)
print(f'Slope = {m}')
b = y2 - m * x2
print(f'Intercept = {b}')
x = symbols('x')
eq = m * x + b
year_of_zero_infection = round(solve(eq, x)[0], 2)
print(f'y = {eq}')
print(f'Infection = 0 after {year_of_zero_infection} years\n')


# Data to graph
x_min = 0
x_max = 30
y_min = 0
y_max = 1000
y_first = m * x_min + b
y_last = m * x_max + b

fig, ax = plt.subplots()
plt.axis([x_min, x_max, y_min, y_max])

ax.set_xlabel("Years since 2004")
ax.set_ylabel("Infected people")
ax.set_title("Linear function")
ax.set_xticks(np.arange(x_min, x_max + 1, 2))
ax.grid(True)

plt.plot([x_min, x_max], [0,0], 'b')
plt.plot([0,0], [y_min, y_max],  'b')
plt.plot([x_min, x_max], [y_first, y_last], label=f'y= {m}x + ({b})')
plt.plot([0], [b], 'ro')
plt.plot([year_of_zero_infection], [0], 'ko', label=f'Year where infections=0')

ax.annotate(f'b= (0, {b})', xy=(0, b), xytext=(0.5, b))
ax.annotate(f'({year_of_zero_infection},0)', xy=(year_of_zero_infection, 0), xytext=(year_of_zero_infection+ 0.5, 0))


ax.legend()
plt.show()



```

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols
from sympy.solvers import solve

# Given points
x1, y1, x2, y2 = 5, 10, 25, 4

# Calculate slope (m) and intercept (b)
m = (y2 - y1) / (x2 - x1)
b = y1 - (m * x1)
print(f'Slope = {m}')
print(f'Intercept = {b}')
eq = m * x + b
zero_profit = round(solve(eq, x)[0], 2)
print(f'y = {eq}')
print(f'Profit was expected to reach 0 in ~ {1980 + zero_profit}\n')

# Graphing
x_min = 0
x_max = 50
y_min = 0
y_max = 15

fig, ax = plt.subplots()
plt.axis([x_min, x_max, y_min, y_max])

ax.set_title('Company profit since 1980')
ax.set_xlabel("Years since 1980")
ax.set_ylabel("Profit in 1000$")
ax.set_xticks(np.arange(x_min, x_max + 1, 2))
ax.set_yticks(np.arange(y_min, y_max + 1, 1))

ax.grid(True)

plt.plot([0, x_max], [b, (m * x_max) + b], label=f'y = {b} - {-m}x')
plt.plot([0], [b], 'go', label='Intercept')
plt.plot([zero_profit], [0], 'ro', label='Zero profit')
plt.annotate(f'b= (0,{b})', xy=(0, b), xytext=(2, 11))
plt.annotate(f'({zero_profit},{0})', xy=(zero_profit, 0), xytext=(zero_profit, 1))


ax.legend()
plt.show()
```
