# Solving Systems

```python
from sympy import *

x, y = symbols('x y')

first = 3 * x - y + 10
second = 4 * x - y

solution = linsolve([first, second], (x, y))
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]
print(f'solution (x, y) = ({x_solution}, {y_solution})')
```

Plot Equations using Sympy library

```python
from sympy import *
from sympy.plotting import plot

var('x y')

first = -2 * x - y + 7
second = 2 * x - y + 8

# Solve the equations using linsolve
solution = linsolve([first, second], (x, y))
x_solution = solution.args[0][0]
y_solution = solution.args[0][1]
print(f'solution (x, y) = ({x_solution}, {y_solution})')

# To graph each equation, solve each one for y
first_eq = Eq(first, 0)
second_eq = Eq(second, 0)

y1 = solve(first_eq, y)[0]
y2 = solve(second_eq, y)[0]

print(f'y1= {y1}, y2= {y2}')

x = symbols('x')
plot(y1, y2, (x, -10, 10))
```
